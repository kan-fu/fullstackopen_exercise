```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(2,2)
x = np.linspace(-10,10,100)
plt.tight_layout()
for i in [0,1]:
    for j in [0,1]:
        k, b = i * 2 - 1, j * 2 - 1
        y = k*x + b
        ax[i][j].set_title(f'k={k} b={b}')
        ax[i][j].set_xlim(-5,5)
        ax[i][j].set_ylim(-5,5)
        ax[i][j].plot(x,y)
        ax[i][j].axhline(y=0, color='k')
        ax[i][j].axvline(x=0, color='k')
```
